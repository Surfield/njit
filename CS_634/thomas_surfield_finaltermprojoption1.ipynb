We will start with the libray imports below:

```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

Next we will create a function to read the divorce.csv data file

```python
def read_csv_data(filename):
    return pd.read_csv('divorce.csv', delimiter=';')
```

Next we will create a function to shuffle and split the data into a train and test set for cross validation

```python
def shuffle_split_data(df):
    df = df.sample(frac=1)
    df_train, df_test = train_test_split(df, test_size=0.2)
    return df_train, df_test
```

Below we run the shuffle_split_data function to shuffle and split the pandas dataframe into test and train datasets. Then we check the data to ensure it was split and shuffled correctly with 20% of the data being moved into the blind test set. We use the head function of the dataframe object to return a subset of the data for investgation. 

```python
df_train, df_test = shuffle_split_data(read_csv_data('divorce.csv'))
```

The Random Forest model is used below, we fit the model with the train features and labels using iloc to grab the approiate columns. Then we run the trained model on the test featutes and labels using the score method which returns the accuracy of the model 

```python
rf_clf = RandomForestClassifier(n_estimators=10).fit(df_train.iloc[:,0:54], df_train.iloc[:,-1])
rf_clf.score(df_test.iloc[:,0:54], df_test.iloc[:,-1]) 
```

```python
df_clf = DecisionTreeClassifier().fit(df_train.iloc[:,0:54], df_train.iloc[:,-1])
df_clf.score(df_test.iloc[:,0:54], df_test.iloc[:,-1]) 
```
